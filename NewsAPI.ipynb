```python
# Dependencies
import tweepy
import numpy as np
import pandas as pd
import time
from datetime import datetime
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')

# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Twitter API Keys
from config import (consumer_key, 
                    consumer_secret, 
                    access_token, 
                    access_token_secret)

# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
# List of Target Users
target_users = ["@BBCWorld", "@CBSNews", "@CNN", "@FoxNews", "@nytimes"]

# Variables for holding sentiments
comp_list = []
pos_list = []
neg_list = []
neu_list = []

date = []
tweets_ago = []
news_outlets = []
text = []

# Looping through all the users
for target_user in target_users:
    # Tweet count
    counter = 0
    

    # Get the last 100 tweets
    for x in range(5):
        # Get all tweets from home feed
        public_tweets = api.user_timeline(target_user, page=x)
        

        # Loop through all the tweets
        for tweet in public_tweets:
            
            # Increase counter
            counter += 1
    
            # Adding values to list
            date.append(tweet['created_at'])
            tweets_ago.append(counter)
            news_outlets.append(target_user)
            text.append(tweet['text'])
            
    
            # Run Vader Analysis on each tweet
            compound = analyzer.polarity_scores(tweet["text"])["compound"]
            pos = analyzer.polarity_scores(tweet["text"])["pos"]
            neu = analyzer.polarity_scores(tweet["text"])["neu"]
            neg = analyzer.polarity_scores(tweet["text"])["neg"]
            
            # Add each value to the sentiment lists
            comp_list.append(compound)
            pos_list.append(pos)
            neu_list.append(neu)
            neg_list.append(neg)
            
            
```

```python
news_data = pd.DataFrame({"News Outlet": news_outlets, "Date of Tweet": date, "Tweets Ago": tweets_ago,
                         "Text": text, "Positive Score": pos_list, "Neutral Score": neu_list,
                         "Negative Score": neg_list, "Composite Score": comp_list})
news_data
```

```python
output = "NewsAPITweets.csv"
news_data.to_csv(output, index = False, header = True)
```

```python
bbc = news_data.loc[news_data["News Outlet"] == "@BBCWorld"]
cbs = news_data.loc[news_data["News Outlet"] == "@CBSNews"]
cnn = news_data.loc[news_data["News Outlet"] == "@CNN"]
fox = news_data.loc[news_data["News Outlet"] == "@FoxNews"]
nyt = news_data.loc[news_data["News Outlet"] == "@nytimes"]

# Create scatter plot
plt.scatter(bbc['Tweets Ago'], bbc['Composite Score'], color = "lightskyblue", edgecolor="black", s=100, alpha=0.95, label='BBC')
plt.scatter(cbs['Tweets Ago'], cbs['Composite Score'], color = "green", edgecolor="black", s=100, alpha=0.95, label='CBS')
plt.scatter(cnn['Tweets Ago'], cnn['Composite Score'], color = "red", edgecolor="black", s=100, alpha=0.95, label='CNN')
plt.scatter(fox['Tweets Ago'], fox['Composite Score'], color = "blue", edgecolor="black", s=100, alpha=0.95, label='Fox')
plt.scatter(nyt['Tweets Ago'], nyt['Composite Score'], color = "yellow", edgecolor="black", s=100, alpha=0.95, label='New York Times')

# Plot title and lables
plt.title("Sentiment Analysis of Media Tweets (%s)" % (time.strftime("%m/%d/%Y")))
plt.xlabel("Tweets Ago")
plt.xlim(105,-5)
plt.ylabel("Tweet Polarity")
plt.ylim(-1.1,1.1)
plt.legend(bbox_to_anchor=(1, 0.95), title='News Outlets')

#Save figure
plt.savefig("Sentiment_Analysis_Tweets.png", bbox_inches="tight")
plt.show()
```

```python
#Gather average composite score
avgbbc = bbc['Composite Score'].mean()
avgcbs = cbs['Composite Score'].mean()
avgcnn = cnn['Composite Score'].mean()
avgfox = fox['Composite Score'].mean()
avgnyt = nyt['Composite Score'].mean()

#Create lists for bar chart
data = [avgbbc, avgcbs, avgcnn, avgfox, avgnyt]
labels = ["BBC", "CBS", "CNN", "Fox", "NYT"]
x_axis = np.arange(len(labels))
colors = ["lightskyblue", "green", "red", "blue", "yellow"]
```

```python
#Set up axes
plt.bar(x_axis, data, color=colors, align="center")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, labels)

#Label axes and plot
plt.title("Overall Media Sentiment based on Twitter (%s)" % (time.strftime("%m/%d/%Y")))
plt.xlabel("News Outlets")
plt.ylabel("Twitter Polarity")

#Save figure
plt.savefig("Overall Media Sentiment.png")
plt.show()
```
